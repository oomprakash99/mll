```python
#read Cleveland Heart Disease data 
condition = pd.read_csv(r'/content/drive/MyDrive/mllabpg/lab5/heart_cleveland_upload.csv')
condition = condition.replace('?',np.nan)
```

```python
#Model Bayesian Network  
model=BayesianModel([('age','trestbps'),('age','fbs'), ('sex','trestbps'),('exang','trestbps'),('trestbps','condition'),('fbs','condition'),('condition','restecg'), ('condition','thalach'),('condition','chol')]) 
```

```python
#Learning CPDs using Maximum Likelihood Estimators
print('\n Learning CPD using Maximum likelihood estimators') 
model.fit(condition,estimator=MaximumLikelihoodEstimator)
```

```python
# Inferencing with Bayesian Network 
print('\n Inferencing with Bayesian Network:') 
heart_infer = VariableElimination(model) 
```

```python
#computing the Probability of HeartDisease given Age
print('\n 1. Probability of HeartDisease given Age=30') 
q=heart_infer.query(variables=['condition'],evidence ={'age':69}) 
print(q)
```

```python
#computing the Probability of HeartDisease given cholesterol 
print('\n 2. Probability of HeartDisease given cholesterol=100') 
q=heart_infer.query(variables=['condition'],evidence ={'chol':200}) 
print(q) 
```
